```
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL Base
url = 'https://books.toscrape.com'
response = requests.get(url)

# Analizar HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Extraer títulos de libros
titulos = [h3.find('a')['title'] for h3 in soup.find_all('h3')]

# Extraer precios
precios = [p.text for p in soup.find_all('p', class_='price_color')]

# Crear dataframe
df = pd.DataFrame({
    'Titulo': titulos,
    'Precio': precios
})

print(df)
```

```
# URL de la página principal
BASE_URL = "https://books.toscrape.com/"
```

```
# Función para extraer datos de libro
def get_book_data(book_url):
    response = requests.get(book_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    title = soup.find("h1").text.strip() # Nombre del libro Etiqueta h1
    price = soup.find("p", class_="price_color").text.strip() # Precio etiqueta p.price_color
    description_tag = soup.find("meta", attrs={"name": "description"})
    description = description_tag["content"].strip() if description_tag else "No description available"

    genre_tag = soup.find("ul", class_="breadcrumb").find_all("li")[2]
    genre = genre_tag.text.strip()

    return {"Title":title, "Price":price, "Genre": genre, "Description": description}
```

```
def get_books_from_page(url):
    # Obtiene la lista de libros de una página específica
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    books = soup.find_all("article", class_="product_pod")

    book_data = []
    for book in books:
        book_link = book.find("h3").find("a")["href"]
        book_url = BASE_URL + "catalogue/" + book_link.replace("../../../","")
        book_data.append(get_book_data(book_url))

    return book_data
```

```
def scrape_all_books():
    # Recorre todas las páginas de libros y extrae los datos
    page = 1
    all_books = []
    while True:
        url = f"{BASE_URL}catalogue/page-{page}.html"
        response = requests.get(url)
        if response.status_code != 200:
            break  # Si la página no existe, terminamos

        all_books.extend(get_books_from_page(url))
        print(f"Scraped page {page}")
        page += 1

    return all_books
```

```
if __name__ == "__main__":
    books_data = scrape_all_books()
    df = pd.DataFrame(books_data)
    df.to_csv("books_data.csv", index = False)
    print("Datos gurdados en books_data.csv")
```

```
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
```

```
import pandas as pd
```

```
def remove_stopwords(description):
    if pd.isna(description) or not isinstance(description, str):
        return description

    words = description.lower().split()
    filtered_words = [word for word in words if word not in stop_words]
    return " ".join(filtered_words)


df['Description'] = df['Description'].apply(remove_stopwords)
```

```
from textblob import TextBlob

def get_polarity(text):
    if isinstance(text, str):
        return TextBlob(text).sentiment.polarity
    return 0

# Aplicar la función
df['Polarity'] = df['Description'].apply(get_polarity)


def get_sentiment(polarity):
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
df['Sentiment'] = df['Polarity'].apply(get_sentiment)
```

```
import matplotlib.pyplot as plt
df['Sentiment'].value_counts().plot(kind='bar')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('No Libros')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```
sentiment_by_genre = df.groupby(['Genre', 'Sentiment']).size().unstack(fill_value = 0)

sentiment_by_genre.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Análisis de sentimeinto por género')
plt.xlabel('Género')
plt.ylabel('No Libros')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
