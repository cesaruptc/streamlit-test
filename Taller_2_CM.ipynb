# 2. Filtrado avanzado

a. Pasajeros de 1ª clase que sobrevivieron mayores de 30 años

```python
# Filtrar los pasajeros que cumplen Pclass = 1, survived = 1, y mayores a 30 años
filtro_a = df[(df['Pclass'] == 1) & (df['Survived'] == 1) & (df['Age'] > 30)]
filtro_a
```

b. Mujeres en 3ª clase con tarifa menor a $20

```python
# Filtro que sea mujer, clase = 3, y tarifa menor a 20
filtro_b = df[(df['Sex'] == 'female') & (df['Pclass'] == 3) & (df['Fare'] < 20)]
filtro_b
```

```python
# Filtrar filas donde la columna name contiene el "Miss"
filtro_c = df[df['Name'].str.contains("Miss", case=False)]    # Se ignoran mayúsculas y minúsculas con false
filtro_c
```

# 3. Transformación de Datos

a. Crea columna "FamilySize" (SibSp + Parch)

```python
df['FamilySize'] = df['SibSp'] + df['Parch']   # Crear columna que va ser igual al número de hermanos/esposos  y número de padres/hijos
df[['SibSp', 'Parch', 'FamilySize']].head()
```

b. Clasifica edades en grupos: Niño(<12), Adolescente(12-17), Adulto(18-65), Anciano(>65)

```python
# Primero crear una función que permita clasificar las edades
def clasificar_edad(edad):
    if pd.isnull(edad):
        return 'Desconocido'
    elif edad < 12:
        return 'Niño'
    elif edad < 18:
        return 'Adolescente'
    elif edad <= 65:
        return 'Adulto'
    else:
        return 'Anciano'

# Usar la función en la columna Age y crear una columna con esto
df['AgeGroup'] = df['Age'].apply(clasificar_edad)
df[['Age', 'AgeGroup']].head(10)
```

c. Convierte "Sex" a numérico (male:0, female:1)

```python
# Convertir sex a númerico con male 0 y female 1

df['SexNum'] = df['Sex'].map({'male': 0, 'female': 1})
df[['Sex', 'SexNum']].head()
```

# 4. Manejo de Valores Faltantes

a. Rellena "Age" con la mediana por clase

```python
# Usar transformación condicional
# Por cada clase se usa la mediana de edad correspondiente
df['Age'] = df['Age'].fillna(df.groupby('Pclass')['Age'].transform('median'))

df['Age'].isnull().sum()
```

b. Elimina filas con "Embarked" faltante

```python
df = df.dropna(subset=['Embarked']) # Donde está nulo embarked eliminar

df['Embarked'].isnull().sum()
```

c. Rellena "Cabin" con "Unknown"

```python
df['Cabin'] = df['Cabin'].fillna("Unknown")

df['Cabin'].isnull().sum()
```
